```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from google.colab import files
```

```python
# 
uploaded = files.upload()
uploaded_file_name = next(iter(uploaded))
im = cv2.imread(uploaded_file_name)
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
# RGB
img_array = np.asarray(im_rgb) #numpy
print(img_array.shape) # (500, 500, 3)3RGB
```

```python
# RGB
print(img_array[1,1,:])
```

```python
im_r = im_rgb.copy()
im_r[:,:, (1,2)]=0 #1,2(GB)
plt.imshow(im_r)
```

```python
im_g = im_rgb.copy()
im_g[:,:, (0,2)]=0 #0,2(RB)
plt.imshow(im_g)
```

```python
im_b = im_rgb.copy()
im_b[:,:, (0,1)]=0 #0,1(RG)
plt.imshow(im_b)
```

```python
# GreenBlue
plt.imshow( im_g + im_b)
```

```python
# GreenRed
plt.imshow( im_r + im_b)
```

```python
# BlueRed
plt.imshow( im_r + im_g)
```
