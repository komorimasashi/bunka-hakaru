```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
from google.colab import files
```

```python
# 画像の読み込み
uploaded = files.upload()
uploaded_file_name = next(iter(uploaded))
im = cv2.imread(uploaded_file_name)
im_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
# RGB画像の中身を見てみよう
img_array = np.asarray(im_rgb) #numpyで扱える配列をつくる
print(img_array.shape) # 配列の次元と要素を調べる　(500, 500, 3)の3がRGBの色チャネル
```

```python
# 左上の画素のRGB値を見てみる
print(img_array[1,1,:])
```

```python
im_r = im_rgb.copy()
im_r[:,:, (1,2)]=0 #全ピクセルの色データの1,2番目(GとB)の数値を０にする
plt.imshow(im_r)
```

```python
im_g = im_rgb.copy()
im_g[:,:, (0,2)]=0 #全ピクセルの色データの0,2番目(RとB)の数値を０にする
plt.imshow(im_g)
```

```python
im_b = im_rgb.copy()
im_b[:,:, (0,1)]=0 #全ピクセルの色データの0,1番目(RとG)の数値を０にする
plt.imshow(im_b)
```

```python
# GreenとBlueのチャネルを足し合わせる
plt.imshow( im_g + im_b)
```

```python
# GreenとRedのチャネルを足し合わせる
plt.imshow( im_r + im_b)
```

```python
# BlueとRedのチャネルを足し合わせる
plt.imshow( im_r + im_g)
```
